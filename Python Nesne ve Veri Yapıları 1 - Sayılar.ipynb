## Değişken Atama
Sayılarla işlem yapmayı öğrendik. Şimdi ise bu sayılara isim vermeyi, yani değişken atama işlemi yapmayı öğrenelim!

<div class="alert alert-block alert-success">
<b>Bilgi:</b> Değişken atama, Veri Bilimi alanında en fazla kullanılan operasyonlardan biridir. 
</div>

Değişken atama için **"=" (eşittir)** sembolünü kullanacağız.

````python
#a isimli bir değişken oluşturup, 5 sayısını atayalım
a = 5
````

Eğer **a** değişkenimi çağıracak olursak, Python bu değişkene (harfe) 5 sayısı muamelesi yapacak!

````python
a
````

````python
#Değişkenleri toplayalım
a + a
````

Peki **a** değişkenine başka bir sayı atadığımızda ne oluyor? Python buna izin veriyor mu?

````python
#Atanan sayıyı değiştirelim
a = 10
````

````python
#Kontrol edelim
a
````

````python
#Oluşturulan değişkeni kullanarak tekrar aynı değişken üzerinde işlem yapalım
a = a + a
````

````python
#Kontrol edelim
a
````

<div class="alert alert-block alert-danger">
<b>Değişken İsimlendirme Uyarıları:</b>
</div>

1. İsimler bir **sayı ile başlayamaz**, harfle başladıktan sonra kullanılabilir.
2. İsimlerde **boşluk (space)** olamaz onun yerine alt çizhgi "_" kullanılabilir.
3. Bu sembollerin hiçbiri kullanılamaz **:'",<>/?|\()!@#$%^&*~-+**
4. Literatürde tamamının **küçük harflerden** oluşması yönünde bir tavsiye mevcut.
5. **'l'** (küçük L harfi), **'O'** (büyük O harfi) veya **'I'** (büyük I harfi) ile tek harf kullanılacak şekilde değişken isimleri vermekten kaçınılmalı.
6. Python için anlam ifade eden **"list"** ve **"str"** gibi isimler koymaktan kaçınılmalı.

<div class="alert alert-block alert-warning">
<b>Egzersiz:</b>
</div>

100 birim olan maaşımızdan vermemiz gereken vergi miktarının %10 olduğunu farz edelim.
- maas
- vergi
- kalan_maas

<br>İsimli değişkenlerimizi ilgili sayıları atayarak **kalan_maas** değişkenimizi hesaplayalım </br>

````python
#Çözüm
#Maaş
maas = 100
#Vergi
vergi = maas*0.1
#Kalan Maaş
kalan_maas = maas-vergi
kalan_maas
````
